```python
# column from series
sum_column=pd.Series([5,5,5,5,5])
car_sales["seats"]=sum_column
car_sales["seats"]=car_sales["seats"].fillna(car_sales["seats"].mean())
car_sales
```

```python
colours_column=pd.Series([5,4,3,2,1])
car_sales["col"]=colours_column
car_sales
```

```python
car_sales["col"].fillna(6, inplace=True)
car_sales
```

```python
#using python
fuel_economy=[32,22,34,53,11,23,14,35,12,43]
car_sales["fuel_economy"]=fuel_economy
car_sales
```

```python
car_sales["total_fuel_used"]=car_sales["Odometer (KM)"]/car_sales["fuel_economy"]
car_sales["fuel_price"]=car_sales["total_fuel_used"]*3.0
car_sales["road_safety_passed"]=True
car_sales
```

```python
car_sales=car_sales.drop("fuel_price", axis=1)
car_sales
```
